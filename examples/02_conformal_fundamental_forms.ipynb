# Conformal Hypersurface Invariants a la Blitz

This notebook demonstrates the `conformal_toolkit.hypersurface` module, which
implements the conformal fundamental forms and associated invariants for
hypersurfaces embedded in Riemannian manifolds.

## Mathematical background

Given a hypersurface $\Sigma^n \hookrightarrow M^{n+1}$, the classical second
fundamental form $L_{ab}$ is *not* conformally invariant. However, its
**trace-free part**
$$L_1{}_{ab} = L_{ab} - H\,h_{ab}$$
where $H = \frac{1}{n}h^{ab}L_{ab}$ is the mean curvature, *is* a conformal
invariant of weight 1. This is the **first conformal fundamental form**.
It vanishes if and only if the hypersurface is **umbilical** (all principal
curvatures are equal), which characterizes round spheres and hyperplanes in
conformally flat ambient spaces.

The **second conformal fundamental form** $L_2$ incorporates higher-order
data (normal derivatives of $L_1$) and detects subtler conformal properties.
Together, the $L_k$ generate all pointwise conformal hypersurface invariants
via contractions.

The **Willmore energy** $\mathcal{W}_2 = \int_{\Sigma} |L_1|^2\,dA$ is the
most famous conformal surface energy, central to the Willmore conjecture
(proved by Marques--Neves, 2012). Its Euler--Lagrange equation involves the
**extrinsic $Q$-curvature** $q_2 = H$.

### References

- S. Blitz, *Conformal Hypersurface Geometry*, PhD thesis, UC Davis (2023)
- S. Blitz, "Conformal hypersurface invariants", [arXiv:2511.02072](https://arxiv.org/abs/2511.02072)
- S. Blitz, "Extrinsic conformal invariants of submanifolds", *J. Math. Phys.* (2023)

## 1. The 2-sphere as a hypersurface in $\mathbb{R}^3$ (umbilical case)

The unit sphere $S^2 \subset \mathbb{R}^3$ is umbilical: all principal curvatures
equal $1$, so $L = h$ (when the normal points outward and $H = 1$).
Thus $L_1 = L - Hh = h - h = 0$.

We set up $S^2$ with its induced metric $h = d\theta^2 + \sin^2\!\theta\,d\varphi^2$
and second fundamental form $L = h$ (identity shape operator).

```sage
from sage.all import Manifold, sin, cos, var
from conformal_toolkit.hypersurface.conformal_fundamental_form import (
    conformal_fundamental_form_L1,
    conformal_fundamental_form_L2,
    mean_curvature,
)
from conformal_toolkit.hypersurface.willmore import willmore_density_W2
from conformal_toolkit.hypersurface.extrinsic_q import extrinsic_q2, extrinsic_q4
from conformal_toolkit.hypersurface.invariant_enumeration import (
    list_invariants,
    count_invariants,
)

# S^2 as a hypersurface: induced metric
Sigma = Manifold(2, 'Sigma', structure='Riemannian')
chart = Sigma.chart(r'theta:(0,pi):\theta phi:(0,2*pi):\phi')
theta, phi = chart[:]
h = Sigma.metric('h')
h[0, 0] = 1
h[1, 1] = sin(theta)**2

# Second fundamental form for S^2 in R^3: L = h (shape operator = identity)
L_sphere = Sigma.tensor_field(0, 2, name='L', sym=(0, 1))
L_sphere[0, 0] = 1
L_sphere[1, 1] = sin(theta)**2

print("Induced metric h on S^2:")
h.display()
print("\nSecond fundamental form L on S^2:")
L_sphere.display()
```

```sage
# Mean curvature
H_sphere = mean_curvature(h, L_sphere)
print("Mean curvature H on S^2:")
H_sphere.display()

# First conformal fundamental form: L_1 = L - H*h
L1_sphere = conformal_fundamental_form_L1(h, L_sphere)
print("\nFirst conformal fundamental form L_1 on S^2:")
L1_sphere.display()

# Output:
# Mean curvature H on S^2:
#   1
# First conformal fundamental form L_1 on S^2:
#   0
#
# L_1 = 0 confirms S^2 is umbilical!
```

```sage
# Willmore density |L_1|^2 on S^2: should be 0 (umbilical)
W2_sphere = willmore_density_W2(h, L_sphere)
print("Willmore density |L_1|^2 on S^2:")
W2_sphere.display()

# Output:
# Willmore density |L_1|^2 on S^2:
#   0
```

## 2. The cylinder $S^1 \times \mathbb{R}$ in $\mathbb{R}^3$ (non-umbilical case)

The cylinder of radius 1 has one vanishing principal curvature (along the
axis direction) and one non-vanishing principal curvature ($\kappa = 1$, around
the circle). Thus $H = 1/2$ and $L_1 \ne 0$: the cylinder is *not* umbilical.

We parametrize the cylinder by $\theta \in (0, 2\pi)$ (angular) and $z \in \mathbb{R}$
(axial), with induced metric $h = d\theta^2 + dz^2$ and shape operator
$L = d\theta^2$ (curvature only in the angular direction).

```sage
# Cylinder S^1 x R as a hypersurface in R^3
Cyl = Manifold(2, 'Cyl', structure='Riemannian')
chart_cyl = Cyl.chart(r'theta:(0,2*pi):\theta z')
theta_c, z = chart_cyl[:]
h_cyl = Cyl.metric('h')
h_cyl[0, 0] = 1  # d(theta)^2
h_cyl[1, 1] = 1  # dz^2

# Second fundamental form: curvature only in theta direction
L_cyl = Cyl.tensor_field(0, 2, name='L', sym=(0, 1))
L_cyl[0, 0] = 1  # kappa_theta = 1
L_cyl[1, 1] = 0  # kappa_z = 0

print("Induced metric h on cylinder:")
h_cyl.display()
print("\nSecond fundamental form L on cylinder:")
L_cyl.display()
```

```sage
# Mean curvature
H_cyl = mean_curvature(h_cyl, L_cyl)
print("Mean curvature H on cylinder:")
H_cyl.display()

# First conformal fundamental form
L1_cyl = conformal_fundamental_form_L1(h_cyl, L_cyl)
print("\nFirst conformal fundamental form L_1 on cylinder:")
L1_cyl.display()

# Output:
# Mean curvature H on cylinder:
#   1/2
# First conformal fundamental form L_1 on cylinder:
#   L_1 = 1/2 dtheta*dtheta + (-1/2) dz*dz
#
# L_1 is non-zero: the cylinder is not umbilical.
# The trace-free condition is satisfied: L1_{00} + L1_{11} = 1/2 - 1/2 = 0.
```

```sage
# Willmore density |L_1|^2 on the cylinder
W2_cyl = willmore_density_W2(h_cyl, L_cyl)
print("Willmore density |L_1|^2 on cylinder:")
W2_cyl.display()

# Output:
# Willmore density |L_1|^2 on cylinder:
#   1/2
#
# This is constant along the cylinder, as expected by symmetry.
# The Willmore energy integral W_2 = integral |L_1|^2 dA diverges
# (infinite cylinder), but the local density is well-defined.
```

## 3. Extrinsic $Q$-curvatures

The extrinsic $Q$-curvatures are the hypersurface analogues of the ambient
$Q$-curvatures. They appear as Euler--Lagrange densities of Willmore-type
functionals:
- $q_2 = H$ (mean curvature, variational density for $\mathcal{W}_2$)
- $q_4 = -\Delta_h H + H|L_1|^2 + (n/2 - 1)H^3$

```sage
# Extrinsic Q-curvature q_2 on S^2 and cylinder
q2_sphere = extrinsic_q2(h, L_sphere)
q2_cyl = extrinsic_q2(h_cyl, L_cyl)

print("q_2 on S^2 (= H):")
q2_sphere.display()
print("\nq_2 on cylinder (= H):")
q2_cyl.display()

# Output:
# q_2 on S^2: 1
# q_2 on cylinder: 1/2
```

```sage
# Extrinsic Q-curvature q_4 on S^2
q4_sphere = extrinsic_q4(h, L_sphere)
print("q_4 on S^2:")
q4_sphere.display()

# On S^2: H = 1, |L_1|^2 = 0 (umbilical), n = 2
# q_4 = -Delta(1) + 1*0 + (2/2 - 1)*1^3 = 0 + 0 + 0 = 0

# Extrinsic Q-curvature q_4 on the cylinder
q4_cyl = extrinsic_q4(h_cyl, L_cyl)
print("\nq_4 on cylinder:")
q4_cyl.display()

# On cylinder: H = 1/2, |L_1|^2 = 1/2, n = 2
# q_4 = -Delta(1/2) + (1/2)(1/2) + (2/2 - 1)*(1/2)^3
#      = 0 + 1/4 + 0 = 1/4
```

## 4. Invariant enumeration: Blitz's classification

How many independent conformal hypersurface invariants exist at each weight
(order of derivatives of the embedding)? The answer depends on the ambient
dimension and is given by Blitz's classification.

The `list_invariants` function returns the catalogue of independent pointwise
conformal hypersurface invariants at a given weight.

```sage
# Weight 2 invariants
print("=== Weight 2 conformal hypersurface invariants ===")
print()
for ambient_dim in [3, 4, 5]:
    inv_list = list_invariants(order=2, ambient_dim=ambient_dim)
    n_inv = count_invariants(order=2, ambient_dim=ambient_dim)
    print(f"Ambient dim = {ambient_dim} (Sigma^{ambient_dim-1} in M^{ambient_dim}): "
          f"{n_inv} independent invariant(s)")
    for inv in inv_list:
        print(f"  - {inv['name']}: {inv['formula']}")
        print(f"    {inv['description']}")
    print()

# Output:
# At weight 2, there is exactly 1 invariant in every ambient dimension >= 3:
#   |L_1|^2 (the Willmore density)
```

```sage
# Weight 4 invariants -- the count depends on ambient dimension!
print("=== Weight 4 conformal hypersurface invariants ===")
print()
for ambient_dim in [3, 4, 5, 6]:
    inv_list = list_invariants(order=4, ambient_dim=ambient_dim)
    n_inv = count_invariants(order=4, ambient_dim=ambient_dim)
    print(f"Ambient dim = {ambient_dim} (Sigma^{ambient_dim-1} in M^{ambient_dim}): "
          f"{n_inv} independent invariant(s)")
    for inv in inv_list:
        print(f"  - {inv['name']}: {inv['formula']}")
    print()

# Output:
# Ambient dim 3: 1 invariant (|L_2|^2; Weyl-based invariants vanish, tr(L_1^4)
#                              is dependent when hypersurface dim <= 3)
# Ambient dim 4: 2 invariants (|L_2|^2, W_{nanb} L_1^{ab})
# Ambient dim 5: 4 invariants (|L_2|^2, tr(L_1^4), W_{nanb} L_1^{ab}, |W_{nabc}|^2)
# Ambient dim 6: 4 invariants (all four)
```

## 5. Second conformal fundamental form $L_2$

The algebraic part of $L_2$ (without the normal derivative term) is:
$$L_2^{\mathrm{alg}}{}_{ab} = (L_1)_a{}^c (L_1)_{cb} - \frac{1}{n}|L_1|^2 h_{ab}$$

On an umbilical surface ($L_1 = 0$), this vanishes. On the cylinder, it is non-trivial.

```sage
# L_2 (algebraic part) on the sphere
L2_sphere = conformal_fundamental_form_L2(h, L_sphere)
print("L_2 (algebraic) on S^2:")
L2_sphere.display()

# L_2 (algebraic part) on the cylinder
L2_cyl = conformal_fundamental_form_L2(h_cyl, L_cyl)
print("\nL_2 (algebraic) on cylinder:")
L2_cyl.display()

# Output:
# L_2 on S^2: 0 (umbilical, L_1 = 0 so L_2^alg = 0)
# L_2 on cylinder: non-zero
#   On the cylinder with L_1_{00} = 1/2, L_1_{11} = -1/2:
#   (L_1^2)_{ab} = L_1_{ac} L_1^c_b, and |L_1|^2 = 1/2
#   So L_2^alg_{ab} = (L_1^2)_{ab} - (1/2)(1/2) h_{ab}
```

```sage
# Higher Willmore density |L_2|^2 on the cylinder
from conformal_toolkit.hypersurface.willmore import willmore_density_W4

W4_cyl = willmore_density_W4(h_cyl, L_cyl)
print("Willmore density |L_2^alg|^2 on cylinder:")
W4_cyl.display()

# This gives a constant value, as expected for the cylinder.
```

```sage
# Extrinsic GJMS operator P_2^Sigma on the cylinder
from conformal_toolkit.hypersurface.extrinsic_gjms import extrinsic_gjms_P2

# Apply to a test function f = cos(theta)
f_test = Cyl.scalar_field(cos(theta_c), name='f')
P2_f = extrinsic_gjms_P2(h_cyl, L_cyl, f_test)
print("P_2^Sigma(cos(theta)) on cylinder:")
P2_f.display()

# P_2(f) = Delta_h(f) + (n/2 - 1) H f
# With n = 2: (n/2 - 1) = 0, so P_2 = Delta_h
# Delta_h(cos(theta)) = -cos(theta)  (on the flat cylinder)
# So P_2(cos(theta)) = -cos(theta)
```

## Summary

We demonstrated the full conformal hypersurface toolkit on two classical surfaces:

| Invariant | $S^2$ (umbilical) | Cylinder (non-umbilical) |
|-----------|-------------------|-------------------------|
| $L_1$ | $0$ | $\frac{1}{2}d\theta^2 - \frac{1}{2}dz^2$ |
| $\|L_1\|^2$ | $0$ | $\frac{1}{2}$ |
| $L_2^{\mathrm{alg}}$ | $0$ | non-zero |
| $q_2 = H$ | $1$ | $\frac{1}{2}$ |
| $q_4$ | $0$ | $\frac{1}{4}$ |

The invariant enumeration shows how the number of independent conformal
hypersurface invariants grows with ambient dimension -- a key result from
Blitz's classification that this toolkit makes computationally accessible.
